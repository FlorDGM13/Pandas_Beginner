##series en pandas
 Es una matriz undiensional que contiene datos de culquier tipo

```python
#Creando una serie en pandas apartir de una lista
a=[1,7,2] #esta es una lista que pasara a ser una serie
myvar=pd.Series(a) #esta es una serie
print(myvar)
```

```python
#acceder a un valor especifico
print(myvar[2])
```

```python
#crear propias etiquetas/indixes
myvar = pd.Series(a, index=["x", "y", "z"])
print(myvar)
```

```python
#acceder a un elemento usando la etiqueta
print(myvar["y"])
```

CREAR UNA SERIE DE PANDAS SIMPLE A PARTIR DE UN DICCIONARIO:

```python
calorias = {"day1": 420, "day2": 380, "day3":390} #este es un diccionario
myvar= pd.Series(calorias)
print(myvar)
```

```python
#creando una serie usando solo datos de day1 y day2
myvar= pd.Series(calorias, index=["day1","day2"])
print(myvar)
```


*   COMBINACIONES DE DATAFRAME



1) CONCATENACION X FILAS

```python
df1 = pd.DataFrame({'NOMBRE': ['Jose', 'Max'],
                    'CARRERA': ['Economia', 'Arquitectura'],
                    'EDAD': [23,26]}).set_index('NOMBRE')
```

```python
df2 = pd.DataFrame({'NOMBRE': ['Aurora', 'Maria'],
                    'CARRERA': ['Medicina', 'Informatica'],
                    'EDAD': [22,28]}).set_index('NOMBRE')
```

```python
df = pd.concat([df1,df2])
df
```

2) CONCATENACION X COLUMNAS

```python
df1 = pd.DataFrame({'AUTOS': ['Nissan', 'Ford', 'Audi'],
                    'COLORES': ['Blanco', 'Azul', 'Rojo']}).set_index('AUTOS')
```

```python
df2 = pd.DataFrame({'AUTOS': ['Nissan', 'Ford', 'Audi'],
                    'MODELO': [2018, 2020, 2022]}).set_index('AUTOS')
```

```python
df = pd.concat([df1,df2], axis=1)
df
```

3) MEZCLAR DATAFRAME

```python
df1 = pd.DataFrame({'AUTOS': ['Nissan', 'Ford', 'Audi'],
                    'COLORES': ['Blanco', 'Azul', 'Rojo']})
```

```python
df2 = pd.DataFrame({'AUTOS': ['Toyota', 'Ford', 'Audi'],
                    'MODELO': [2018, 2020, 2022]})
```

```python
df = pd.merge(df1,df2, on = 'AUTOS', how='left') #filas-outer
df
#left solo toma en cuenta df1
```
